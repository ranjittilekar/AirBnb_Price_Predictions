### Make Feature function

Perform steps during feature engineering

```python
# Create Additional Property Features

def make_features(dataframe):

    #---------------------    
    # Fill Missing 'NaN'
    #---------------------

    # Continous - Fill with Estimate: Median
    fill_median = ['accommodates', 
                    'bathrooms', 
#                     'first_review',
#                     'host_response_rate',
#                     'host_since',
#                     'last_review',
                    'number_of_reviews',
                    'review_scores_rating',
                    'bedrooms',
                    'beds']
    
    # Fill with Group Median
#     group_by = ['neighbourhood','bed_type']  
#     dataframe[fill_median] = dataframe[fill_median].fillna(dataframe.groupby(group_by)[fill_median].transform('median'))

    #Fill with Median
    dataframe[fill_median] = dataframe[fill_median].fillna(dataframe[fill_median].median())


    
    # Categorical - Fill with 'Blank' Indicator    
    fill_blank = ['property_type',
                    'room_type',
                    'bed_type',
                    'cancellation_policy',
                    'cleaning_fee',
                    'city',
                    'description',
                    'host_has_profile_pic',
                    'host_identity_verified',
                    'instant_bookable',
                    'name',
                    'neighbourhood',
                    'thumbnail_url', 
                    'zipcode']
    
    dataframe[fill_blank] = dataframe[fill_blank].fillna('Blank')
              
    #---------------------
    # Engineer Categorical Features
    #---------------------
    
    # Simplify URL link to Yes / No
    dataframe['thumbnail_url'].loc[dataframe['thumbnail_url'].notnull()] = 'Y'
    dataframe['thumbnail_url'].loc[dataframe['thumbnail_url'].isnull()] = 'N'

    #---------------------    
    # Engineer Continuous Features
    #---------------------
   
    
    return dataframe
```

## Preprocess Data

1. Standardize continous variables
2. One-hot Encode categorical variables

```python
def preprocess_data(dataframe):
    from sklearn.preprocessing import MinMaxScaler, StandardScaler
       
    # Standardize the data type
    dataframe[continuous_variables] = dataframe[continuous_variables].astype(float).fillna(dataframe[continuous_variables].median())
    dataframe[categorical_variables] = dataframe[categorical_variables].astype(object)
    
    # Numeric Features: Normalize Scale Numerical Values
    # scaler = MinMaxScaler() # default=(0, 1)
    # dataframe[continuous_variables] = scaler.fit_transform(dataframe[continuous_variables])
    
    # Numeric Features: Normalize Z-Score Numerical Values
    std_scaler = StandardScaler() 
    dataframe[continuous_variables] = std_scaler.fit_transform(dataframe[continuous_variables]).round(3)
      
    # Create array of dummies
    dataframe = pd.get_dummies(dataframe, drop_first=True, dummy_na=True)

    # Display shape
    print ("Complete: The preprocessed data shape: {}".format(dataframe.shape))
    display(dataframe.head())
    
    return dataframe
```

## Deep Learning: Regression using Neural Network

### Training, validation and split
In order to train our deep learning model and evaluate it, it’s important to split the data into training, validation and testing datasets. We will train the model on a training set, and tune the model by validating the results on the validation dataset. Finally we will measure the ability of our model to generalize by testing our model on the testing dataset.

```python
# Set feature and label variable
features = X_train
labels = y_train

# Shuffle the train data order, and split by 60% train, 20% validation, 20% test data
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=104)

X = np.zeros(shape=X_train.shape[0], dtype=int)
y = np.zeros(shape=y_train.shape[0], dtype=int)

splitter = sss.split(X, y)

# Create randomize index
train_index, validation_index = next(splitter)
test_index = validation_index[:int(len(validation_index)/2)]
validation_index = validation_index[int(len(validation_index)/2):]

# Use iloc to look up dataframe index
train_x, train_y = features.iloc[train_index], labels.iloc[train_index]
test_x, test_y = features.iloc[test_index], labels.iloc[test_index]
val_x, val_y = features.iloc[validation_index], labels.iloc[validation_index]

print("\t\t\tFeature Shapes:")
print("Train set: \t\t{}".format(train_x.shape), 
      "\nValidation set: \t{}".format(val_x.shape),
      "\nTest set: \t\t{}".format(test_x.shape))
```

```python
input = train_x.shape[1]

# Original Deep Learning Model
def baseline_model():
	# create model
    model = Sequential()
    model.add(Dense(input, input_dim=input, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
# Transform into np.aarray for tensorflow
train_x = np.array(train_x)
train_y = np.array(train_y)

test_x = np.array(test_x)
test_y = np.array(test_y)

val_x = np.array(val_x)
val_y = np.array(val_y)
```

### Build the Network:
First Input layer: It is important for the first input layer to equal the number of property features.

Hidden layer: The dense layers are used to prevent overfitting to the training set. The structure of the network was built through trial and error, and experiments. Note that the first iterations only had 2-3 layers. Additional layers were added at the end.

Output layer: The linear activation function is used for regression problems with 1 output.
Compile Model: Tested several optimization algorithms (‘sgd’, ‘RMSprop’, and ‘Adam’) and selected based on results. Also tune the learning rate.

Fit Model: Tune the number of epochs and batch size to match the computation capacity on my local computer. Training deep learning models are computationally expense.

```python
input = train_x.shape[1]

# create model
model = Sequential()
model.add(Dense(input, input_dim=input, kernel_initializer='normal', activation='relu'))
model.add(Dense(input, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(input, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(input, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(input, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(input, kernel_initializer='normal', activation='relu'))
model.add(Dense(input, kernel_initializer='normal', activation='relu'))
model.add(Dense(input, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, activation='linear'))

model.summary()
```

## Test Several Optimizer (sgd, rmsp, adam):

```python
# Change Directory to models
os.chdir("{}models".format(rootDirectory))
print(os.getcwd())
```

```python
# Compile model

#SGD: Tune learning rate 
optimizers.SGD(lr=0.01, momentum=0.005, decay=0.0, nesterov=False)
model.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
from keras.callbacks import ModelCheckpoint  

### specify the number of epochs and save best validation result

epochs = 50
checkpointer = ModelCheckpoint(filepath='weights.best.from_scratch_sgd.hdf5', 
                               verbose=1, save_best_only=True)

#train data
sgd_history = model.fit(train_x, train_y, epochs=epochs, batch_size=256, callbacks=[checkpointer], verbose=1, validation_data=(val_x, val_y))
```

```python
#Visualization
# list all data in history
print(sgd_history.history.keys())

# summarize history for loss
fig, ax = plt.subplots(figsize=(14,10))
plt.plot(sgd_history.history['loss'])
plt.plot(sgd_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# RMSPROP
optimizers.RMSprop(lr=0.01, rho=0.9, epsilon=1e-08, decay=0.0)
model.compile(loss='mean_squared_error', optimizer='rmsprop')
```

```python
from keras.callbacks import ModelCheckpoint  

### specify the number of epochs and save best validation result

epochs = 50
checkpointer = ModelCheckpoint(filepath='weights.best.from_scratch_RMSprop.hdf5', 
                               verbose=1, save_best_only=True)
#train data
RMSprop_history = model.fit(train_x, train_y, epochs=epochs, batch_size=256, callbacks=[checkpointer], verbose=1, validation_data=(val_x, val_y))
```

```python
#Visualization
# list all data in history
print(RMSprop_history.history.keys())

# summarize history for loss
fig, ax = plt.subplots(figsize=(14,10))
plt.plot(RMSprop_history.history['loss'])
plt.plot(RMSprop_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#ADAM
optimizers.Adam(lr=0.01, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
from keras.callbacks import ModelCheckpoint  

### specify the number of epochs and save best validation result

epochs = 50
checkpointer = ModelCheckpoint(filepath='weights.best.from_scratch_adam.hdf5', 
                               verbose=1, save_best_only=True)

#train data
adam_history = model.fit(train_x, train_y, epochs=epochs, batch_size=256, callbacks=[checkpointer], verbose=1, validation_data=(val_x, val_y))
```

```python
#Visualization
# list all data in history
print(adam_history.history.keys())

# summarize history for loss
fig, ax = plt.subplots(figsize=(14,10))
plt.plot(adam_history.history['loss'])
plt.plot(adam_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
